```python
columns_to_remove=['tags','dislikes']
df=df.drop(columns=columns_to_remove)
df.shape
```

```python
from datetime import datetime
import pandas as pd
df["trending_date"] = df["trending_date"].apply(lambda x: datetime.strptime(x,'%y.%d.%m'))
print(df.head())
```

```python
df['publish_time']=pd.to_datetime(df['publish_time'])
print(df.head())
```

```python
df['publish_month']=df['publish_time'].dt.month
df['publish_day']=df['publish_time'].dt.day
df['publish_hour']=df['publish_time'].dt.hour
df.head(3)
```

```python
df['year']=df['publish_time'].dt.year
yearly_counts=df.groupby('year')['video_id'].count()
yearly_counts.plot(kind='bar',xlabel='year',ylabel='totle publish count',title='totle publish video per year')
plt.show()
```

```python
category_views=df.groupby('category_id')['views'].sum().reset_index()
top_categories=category_views.sort_values(by='views',ascending=False).head(6)
plt.bar(top_categories['category_id'],top_categories['views'])
plt.xlabel('category_id')
plt.ylabel('totle views')
plt.title('top 6 categories')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='category_id',data=df,order=df['category_id'].value_counts().index)
plt.xticks(rotation=90)
plt.title('video count by category')
plt.show()
```

```python
sns.scatterplot(data=df,x='views',y='likes')
plt.title('views vs dislikes')
plt.xlabel('views')
plt.ylabel('likes')
plt.show()
```

```python
plt.figure(figsize=(26,7))
plt.subplots_adjust(wspace=0.3,hspace=0.4,top=1)
plt.subplot(2,2,1)
g2=sns.countplot(x='comments_disabled',data=df)
g2.set_title("comments disabled",fontsize=20)
plt.subplot(2,2,2)
g2=sns.countplot(x='ratings_disabled',data=df)
g2.set_title("rating disabled",fontsize=20)
plt.subplot(2,2,3)
g2=sns.countplot(x='video_error_or_removed',data=df)
g2.set_title("video error or removed",fontsize=20)
plt.show()
```
